```python
import pandas as pd

# Reading and saving the csv file to a variable.
df = pd.read_csv("./raw_listings.csv")

# Deleting unnecessary data.
df.drop(columns=["id", "name", "host_id", "host_name", "neighbourhood", "latitude", "longitude", "reviews_per_month", "calculated_host_listings_count", "number_of_reviews", "number_of_reviews_ltm", "license" ], inplace=True)
# Deleting rows that have invalid or null values.
df.dropna(axis=0, how="any", inplace=True)
# Deleting duplicate rows.
df.drop_duplicates(keep="first", inplace=True)

# Converting the last_review column to a suitable format so that we can filter it.
df["last_review"] = pd.to_datetime(df["last_review"])
# Filtering old rental houses that haven't been used in over a year.
df = df.loc[df["last_review"] > "2022-06-01"]

# Making the index of the data frame match the number of rows.
df.reset_index(drop=True, inplace=True)

# Creating a new csv file with the cleaned data.
df
```
